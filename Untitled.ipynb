```python
from topograd import genpd
from topograd import major
```

Generating sythetic data below:

```python
import matplotlib.pyplot as plt
import numpy as np
mean = [3.1,0]
cov = [[0.07,0],[0,0.07]] 
x,y = np.random.multivariate_normal(mean,cov,500).T
plt.plot(x,y,'x'); plt.axis('equal'); plt.show()
```

```python
mean = [0,0]
cov = [[0.7,0],[0,0.7]]
x1,y2 = np.random.multivariate_normal(mean,cov,500).T
plt.plot(x,y,'x'); plt.axis('equal')
plt.plot(x1,y2,'x'); plt.axis('equal'); plt.show()
```

```python
finalx = np.append(x1,x)
finaly = np.append(y2,y)
plt.plot(finalx,finaly,'x'); plt.axis('equal')
final = np.zeros([1000,2])
final[:,0] = finalx
final[:,1] = finaly
```

Function genpd:
genpd(pc, k1,k2,tau1,thresh)

pc is the dataset we use, which should be a point cloud in array format.

k1 is the number of neighbors of neighborhood graph.

k2 is the number of k-nearest neighbor used to construct density map, techniquely larger k2 value means larger algorithm affect. 

tau1 is the bandwidth of kernel density estimator.

thresh is threshold for merging connect components (candidate clusters) togather. Larger thershold means less clusters kept. But notice, to see the overall persistence diagram of dataset, we need to set this hyperparameter to be 1.

The clustering result and persistence diagram will be generated by this function.

The overall persistence diagram can be seen as a guidance of clustering.

The overall persistence diagram of synthetic data is shown below.

```python
import time
import matplotlib.pyplot as plt
start = time.time()
X = final
dddd,pd = genpd(X,30,10,0.1,1)
# dddd is cluster label, but when you set thresh to be 1 , ALL the candidate clusters are merged, for this dataset there is only one cluster
# left after merging. pd is the overall persistence diagram. For this dataset we can see there is only 1 salient candidate cluster. If we set
# the thresh between the persistence of most salient and the rest. We can see 2 cluster shown in next. To find this thresh we just need to use
# birth time minus death time to get the persistence time of candidate clusters.
print(np.sort(pd[:,0] - pd[:,1]))
# we can see 0.35791 is so much bigger than the rest (0.0744). So if we set a number between 0.35791 and 0.0744. The corresponding candidate 
# cluster will be kept.
ccc = np.zeros(1000)
k = 0
for x,y in dddd.items():
    for i in y:
        ccc[y] = k
    k = k+1
plt.figure(figsize = (10,10))
plt.scatter(X[:,0],X[:,1],s = 4, c = ccc )
plt.figure(figsize = (10,10))
plt.scatter(pd[:,0],pd[:,1],s = 4 )
plt.plot(np.array([0,1]),np.array([0,1]))
print(time.time() - start)
```

We can see when we set thresh to be 0.07, one more candidate cluster is shown, but it's not very salient on persistence digram. What if we say this cluster is important too, we want it to be salient on persistence diagram, so that easier to recognize. Our major function:
major(pc, k1,k2,tau1, destnum, learning_rate, epoch_num)
is designed for this requirement. (pc, k1, k2, tau1) has the same meaning as genpd. 

destnum here is the number of clusters to be more salient, if destnum = 1 than the most salient candidate cluster will be kept, all the rest candidate clusters will be weaken (corresponding dot on persistence diagram will be pushed toward diagonal until disappear). If destnum = n (n > 1), the most salient n -1 connect components will change to be more close to the most salient one. The other candidate cluster will be weaken. This preocess is shown below.

res is the point cloud after learning.

pd is the overall persistence diagram.

```python
import time
import matplotlib.pyplot as plt
start = time.time()
X = final
res,pd = major(X,30,10,0.1,2,0.01,500)
# pc = X, k1 = 30, k2 = 10, tau1 = 0.1, destnum = 2, lr = 0.01, epoch = 500
# ccc = np.zeros(1000)
plt.figure(figsize = (10,10))
plt.scatter(res[:,0],res[:,1],s = 4 )
plt.figure(figsize = (10,10))
plt.scatter(pd[:,0],pd[:,1],s = 4 )
plt.plot(np.array([0,1]),np.array([0,1]))
print(np.sort(pd[:,0] - pd[:,1]))
print(time.time() - start)
```

we can see after learning, the most salient persistence is 5.62541753e-01, the second salient persistence is 5.46817213e-01, very close to each other, and far from the other candidate cluster (3.60938311e-02)

to verify this we set thresh to be 0.5 we can see 2 more candidate cluster is shown below.

```python
X = res
dddd,pd = genpd(X,30,10,0.1,0.5)
ccc = np.zeros(1000)
k = 0
for x,y in dddd.items():
    for i in y:
        ccc[y] = k
    k = k+1
plt.figure(figsize = (10,10))
plt.scatter(X[:,0],X[:,1],s = 4, c = ccc )
```

k2 is the number of k-nearest neighbor used to construct density map. When we manipulate the persistence diagram, we are actually manipulating the density map directly. So bigger k2 means stronger affect.

k1 is the number of nearest data points, which can be recognized as neighbor. This hyper parameter is important for cluster result.
